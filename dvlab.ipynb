```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df=pd.read_csv("/content/drive/MyDrive/ds_salaries.csv")

# Extract the relevant columns
company_size = df['company_size']
salaries = df['salary']

# Group the data by company size and calculate the mean salary
mean_salaries = df.groupby('company_size')['salary'].mean().reset_index()

# Fit a polynomial regression model
degree = 1  # Degree of the polynomial regression model
coefficients = np.polyfit(mean_salaries.index, mean_salaries['salary'], degree)
poly_function = np.poly1d(coefficients)

# Create the trendline values
trendline_company_size = np.linspace(mean_salaries.index.min(), mean_salaries.index.max(), 100)
trendline_salaries = poly_function(trendline_company_size)

# Calculate confidence intervals
y_hat = poly_function(mean_salaries.index)
residuals = mean_salaries['salary'] - y_hat
MSE = np.sum(residuals ** 2) / (len(mean_salaries.index) - degree - 1)
standard_error = np.sqrt(MSE)
confidence_band = 1.96 * standard_error

# Plot the data points, trendline, and confidence band
plt.scatter(mean_salaries.index, mean_salaries['salary'], label='Data')
plt.plot(trendline_company_size, trendline_salaries, color='r', label='Trendline')
plt.fill_between(trendline_company_size, trendline_salaries - confidence_band, trendline_salaries + confidence_band, color='r', alpha=0.2, label='Confidence Band')
plt.xlabel('company_size')
plt.ylabel('salary')
plt.title('Salary Trend by Company Size')
plt.xticks(mean_salaries.index, mean_salaries['company_size'], rotation='vertical')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
company_size = df['company_size']
salaries = df['salary']

# Group the data by company size and calculate the mean salary
mean_salaries = df.groupby('company_size')['salary'].mean().reset_index()

# Fit a polynomial regression model
degree = 1  # Degree of the polynomial regression model
coefficients = np.polyfit(mean_salaries.index, mean_salaries['salary'], degree)
poly_function = np.poly1d(coefficients)

# Create the trendline values
trendline_company_size = np.linspace(mean_salaries.index.min(), mean_salaries.index.max(), 100)
trendline_salaries = poly_function(trendline_company_size)

# Calculate confidence intervals
y_hat = poly_function(mean_salaries.index)
residuals = mean_salaries['salary'] - y_hat
MSE = np.sum(residuals ** 2) / (len(mean_salaries.index) - degree - 1)
standard_error = np.sqrt(MSE)
confidence_band = 1.96 * standard_error

# Plot the data points, trendline, and confidence band
plt.scatter(mean_salaries.index, mean_salaries['salary'], label='Data', alpha=0.7, edgecolors='black')
plt.plot(trendline_company_size, trendline_salaries, color='r', label='Trendline')
plt.fill_between(trendline_company_size, trendline_salaries - confidence_band, trendline_salaries + confidence_band, color='r', alpha=0.2, label='Confidence Band')

# Add jittering to the scatter plot
x_jitter = np.random.normal(scale=0.07, size=len(mean_salaries.index))
y_jitter = np.random.normal(scale=2000, size=len(mean_salaries.index))
plt.scatter(mean_salaries.index + x_jitter, mean_salaries['salary'] + y_jitter, alpha=0.7, edgecolors='black')

plt.xlabel('company_size')
plt.ylabel('salary')
plt.title('Salary Trend by Company Size')
plt.xticks(mean_salaries.index, mean_salaries['company_size'], rotation='vertical')
plt.legend()
plt.tight_layout()
plt.show()
```
